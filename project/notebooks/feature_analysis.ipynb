### Problem with adding features due to PCA
The dataset used has already undergone PCA so there are 28 anonymized features other than amount and time, which aren't directly related. Therfore adding features that aren't time based is difficult, as the featues have already been engineered.  

```python
#use datetime to group by hour and minute
df['datetime'] = pd.to_datetime(df['Time'], unit='s')
df = df.sort_values('Time').reset_index(drop=True)

# Transactions per hour (group by hour)
df['hour_group'] = df['datetime'].dt.floor('h')
transactions_per_hour = df.groupby('hour_group').size().reset_index(name='hourly_transactions')
df = df.merge(transactions_per_hour, on='hour_group', how='left')

df['minute_group'] = df['datetime'].dt.floor('min')
transactions_per_minute = df.groupby('minute_group').size().reset_index(name='minute_transactions')
df = df.merge(transactions_per_minute, on='minute_group', how='left')

df
```
