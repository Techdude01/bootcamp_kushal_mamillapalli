# Setup: Generate Sample Dataset

This cell creates the required folder structure (`data/raw/` and `data/processed/`) relative to the notebook, and generates the sample CSV dataset with missing values. 
This ensures the dataset is ready for cleaning functions and saves it to `data/raw/sample_data.csv`.

```python
import os
import pandas as pd
import numpy as np
import sys
sys.path.append('..')
from src import cleaning
# Define folder paths relative to this notebook
raw_dir = '../data/raw'
processed_dir = '../data/processed'
data_name = "/fraud_kaggle_creditcard_source-kaggle_dataset-creditcardfraud_20250820-213245"


# Load DataFrame
df = pd.read_parquet(processed_dir + data_name + ".parquet")
```

## Apply Cleaning Functions

```python

df = cleaning.fill_missing_median(df, ['Amount'])
#not any specific column
#very high as dataset looks good
df = cleaning.drop_missing(df, threshold=0.9)
#age and income have numeric values
df = cleaning.normalize_data(df, ['Amount'])
```
